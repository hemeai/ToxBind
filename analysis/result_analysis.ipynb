```python
combined_df = pd.read_csv('./combined_data.csv')
results_ipae_df = pd.read_csv('./results_ipae.csv')
```

```python
merged_df = pd.merge(combined_df, results_ipae_df, left_on='Design', right_on='Design')
merged_df
```

```python
for value in merged_df.columns.to_list(): 
    print(value)
```

```python
specific_columns = [
    'Rank',
    'Design',
    'Length',
    'ipae_score',
    'Average_i_pTM',
    'Target_Hotspot',
    'Sequence',
    'Average_pAE',
    'Average_i_pAE',
    'Average_pTM',
    'Average_pLDDT',
    'Average_i_pLDDT',
    'Average_ss_pLDDT',
    'Average_Target_RMSD',
    'Average_Hotspot_RMSD',
    'Average_Binder_pLDDT',
    'Average_Binder_pTM',
    'Average_Binder_pAE',
    'Average_Binder_RMSD',
    'DesignTime',
    'Notes',
    'TargetSettings'
]
```

```python
sorted_merged_df = merged_df.sort_values(by='ipae_score', ascending=True)
sorted_merged_df = sorted_merged_df[specific_columns]
sorted_merged_df
```

```python
domain_1_df = sorted_merged_df[sorted_merged_df['TargetSettings'] == '6aru_final_chain_A_domain_1']
domain_1_df
```

```python
domain_3_df = sorted_merged_df[sorted_merged_df['TargetSettings'] == '6aru_final_chain_A_domain_3']
domain_3_df = domain_3_df[domain_3_df['Length'] > 100]
```

```python
domain_3_no_clash_df = domain_3_df[domain_3_df['Notes'] != 'Relaxed structure contains clashes.']
domain_3_no_clash_df = domain_3_no_clash_df.sort_values(by=['Average_i_pTM', 'ipae_score', 'Rank'], ascending=[False, False, True])
domain_3_no_clash_df['Design_Batch'] = domain_3_no_clash_df['Design'].str.replace(r'_mpnn\d*$', '', regex=True)
domain_3_no_clash_df = domain_3_no_clash_df.drop_duplicates(subset='Design_Batch', keep='first')
domain_3_no_clash_df
```

```python
domain_3_with_clash_df = domain_3_df[domain_3_df['Notes'] == 'Relaxed structure contains clashes.']
domain_3_with_clash_df
```

```python
top_10_df = pd.concat([domain_1_df, domain_3_with_clash_df.tail(1), domain_3_no_clash_df.head(7) ], ignore_index=True)
top_10_df
```

```python
fasta_file_name = "final_submission_2"
final_fasta_content = ""
# Iterate over each row of the DataFrame
for index, row in top_10_df.iterrows():
    design = row['Design']
    binder_seq_len = row['Length']
    sequence = row['Sequence'] 

    # Create the FASTA file content
    fasta_content = f">{design}\n{sequence}\n\n"
    final_fasta_content += fasta_content

# Write the FASTA file
fasta_file_path = f"{fasta_file_name}.fasta"
with open(fasta_file_path, "w") as fasta_file:
    fasta_file.write(final_fasta_content)
```
